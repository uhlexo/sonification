```python
import midi
import pandas as pd
pd.options.display.float_format = '{:.4f}'.format
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import datetime as dt
%matplotlib inline
from IPython.display import display
```

```python
# the data here has already been cleaned so that it'll work nicely with the midi package
# values need to be integers between 0-127

hourly = pd.read_csv("data/Sonification_Summary_midi.csv")
display(hourly)
```

```python
hourly_list = hourly.values.tolist()

hourly_list
```

```python
# Instantiate a MIDI Pattern (contains a list of tracks)
# Read more about resolution and ticks here:
# https://github.com/vishnubob/python-midi#side-note-what-is-a-midi-tick

pattern = midi.Pattern(resolution=200)

# Instantiate a MIDI Track (contains a list of MIDI events)
track = midi.Track()

# Append the track to the pattern
pattern.append(track)

# Ticks are relative to the previous event
# Below it's set to a constant because the data is sampled at a regular interval with no gaps
cc_changes = []
for d in hourly_list:
    cc_changes.append(midi.ControlChangeEvent(tick=100, channel=0, control=3, value=d[1]))

# Here's the code that generates all the different events that you can control
# https://github.com/vishnubob/python-midi/blob/master/src/events.py

track.extend(cc_changes)

# Add the end of track event, append it to the track
eot = midi.EndOfTrackEvent(tick=100)
track.append(eot)

# Print out the pattern
print pattern
```

```python
# Save the pattern to disk
midi.write_midifile("midi/Sonification_Summary_midi.mid", pattern)
```
